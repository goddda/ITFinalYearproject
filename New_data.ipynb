```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
from sklearn import preprocessing
from scipy.signal import find_peaks
```

```python
data_dir = os.getcwd()
data = pd.read_csv(data_dir + "/accel.csv")
data = pd.read_csv(data_dir + "/accel.csv")
i, mind, maxd = 0, 0, 0
plt.figure(0)
for a in ['accelX','accelY','accelZ']:
#     plt.figure(i)
    d = data[data['ID']==3][a]
    maxd = max(d)
    mind == min(d)
    plt.plot((d-mind)/(maxd-mind))
    i += 1
```
